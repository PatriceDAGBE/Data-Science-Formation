```python
# YouTube_Data_Scraping_Preprocessing_and_Analysis_using_Python
import time
from selenium import webdriver
from bs4 import BeautifulSoup
import xlsxwriter
import pandas as pd
```

```python
urls = [
    'https://www.youtube.com/c/GeeksforGeeksVideos/videos'
]
driver = webdriver.Chrome()
soup = None
```

```python
for url in urls:
    times = 0
    driver.get('{}/videos?view=0&sort=p&flow=grid'.format(url))
    while times < 5:
        time.sleep(1)   
        driver.execute_script(
            "window.scrollTo(0, document.documentElement.scrollHeight);")
        times += 1
    content = driver.page_source.encode('utf-8').strip()
    soup = BeautifulSoup(content, 'lxml')
print("Channel Scraping Done!")
```

```python
#Title
titles_all = soup.find_all(id='video-title')
titles = []
for title in titles_all:
    titles.append(title.text)
print(len(titles))
```

```python
#Views
views = []

#Durations
durations = []

videos = soup.select("ytd-rich-item-renderer")

spans = soup.select('#details #metadata-line span')
for i, span in enumerate(spans):
    if i % 2 == 0:
        views.append(spans[i].get_text(strip=True))

for video in videos:
    badge = video.find("badge-shape")
    if badge:
        dur = badge.find("div", class_="badge-shape-wiz__text")
        if dur:
            durations.append(dur.get_text(strip=True))
```

```python
# Excel file creating

workbook = xlsxwriter.Workbook('file.xlsx')
worksheet = workbook.add_worksheet()

worksheet.write(0, 0, "Title")
worksheet.write(0, 1, "Views")
worksheet.write(0, 2, "Duration")

row = 1
for title, view, dura in zip(titles,views,durations):
    worksheet.write(row, 0, title)
    worksheet.write(row, 1, view)
    worksheet.write(row, 2, dura)
    row += 1

workbook.close()
```

```python
data = pd.read_excel('file.xlsx')
data.head()
```

```python
# Views column cleaning
data['Views'] = data['Views'].str.replace(" views","")

new = []

for i in data['Views']:
    if(i.endswith('K')):
        i = i.replace('K','')
        new.append(float(i) * 1000)
    else :
        new.append(i)

data['Views'] = new
```

```python
#Duration column cleaning
data['Duration'] = data['Duration'].str.replace("\n","")

new2 = []

for i in data['Duration']:
    if(i=='SHORTS' or len(i.split(':'))==1):
        new2.append(i)
    elif(len(i.split(':'))==2):
        i = i.split(':')
        tim = int(i[0])*60 + int(i[1])
        new2.append(tim)
    elif(len(i.split(':'))==3):
        i = i.split(':')
        tim = int(i[0])*3600 + int(i[1])*60 + int(i[2])
        new2.append(tim)
        
data['Duration'] = new2
```

```python
#Duration column categorization
for i in data['Duration'].index:
    val = data['Duration'].iloc[i]
    if(val=='  SHORTS'):
        continue
    elif(val in range(0,900)):
        data.loc[i,'Duration'] = 'Mini-Videos'
    elif(val in range(901,3600)):
        data.loc[i,'Duration'] = 'Long-Videos'
    else:
        data.loc[i,'Duration'] = 'Very-Long-Videos'
data.head()
```

```python
from tqdm import tqdm
import re
import nltk
nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import stopwords

def preprocess_text(text_data):
    preprocessed_text = []
    
    for sentence in tqdm(text_data):
        sentence = re.sub(r'[^\w\s]', '', sentence)
        preprocessed_text.append(' '.join(token.lower()
                                    for token in str(sentence).split()
                                        if token not in stopwords.words('english')))

    return preprocessed_text

print(data["Title"].values)
preprocessed_review = preprocess_text(data['Title'].values)
data['Title'] = preprocessed_review
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
consolidated = ' '.join(word for word in data['Title'].astype(str))
wordCloud = WordCloud(width=1600, height=800, random_state=21,
                      max_font_size=110, collocations=False)
plt.figure(figsize=(15, 10))
plt.imshow(wordCloud.generate(consolidated), interpolation='bilinear')
plt.axis('off')
plt.show()
```
