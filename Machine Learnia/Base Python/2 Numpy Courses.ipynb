````python
def initialisation(m, n):
    # m: nombre de lignes
    # n: nombre de colonnes
    # retourne une matrice aléatoire (m, n+1)
    # avec une colonne biais ("remplie de "1" ") tout à droite
    
    # Fonctions possibles à utiliser
    # reshape, random.randn, concatenate
    
    A = np.random.randn(m, n)
    
    B = np.ones((A.shape[0], 1))
    
    C = np.concatenate((A, B), axis=1)
    
    return C

C = initialisation(4, 2)
C
````

# Exercise

Standardiser la matrice A sur chaque colonne

```txt
    A−mean(Acolonne)
A = ________________
    std(Acolonne)
```
    
Résultat: chaque colonne doit avoir moyenne = 0 et ecart type = 1

````python
np.random.seed(0)

A = np.random.randint(0, 100, [10, 5])
A
````

````python
# La standardisation est très pratique et très courante en Machine Learning et en Deep Learning
# Elle permet de mettre sur même échelle toutes les différentes lignes ou toutes les différentes
# colonnes de notre dataset

# Pour cela, il existe plusieurs techniques qui nous permettent de faire cela
# On peut les retrouver suivant ce lien en dessous:
#=========================================================
# https://en.wikipedia.org/wiki/Feature_scaling
#=========================================================
````

````python
D = (A - A.mean(axis=0)) / A.std(axis=0)
D
````

````python
D.mean(axis=0) ## Le résultat donne des valeurs très proches de 0
````

````python
D.std(axis=0) ## l'écart-type est 1
````

````python
# Cela voudrait dire que chaque de notre colonne suive une distribution normale
# standar absolument parfaite.
# Et avoir des données qui suivent ce genre de distribution, eh bien, çà facilite énormément
# l'apprentissage statistics tout simplement parce que les données sont très bien réparties
````

## BROADCASTING

Le broadcasting c'est un programme qui a été conçu pour faciliter les calculs entre les deux matrices différentes

````python
A = np.random.randint(0, 10, [2, 3])
B = np.ones((2, 3))
A
````

````python
B
````

````python
A + B
````

````python
#NB: On peut faire ces genres d'opérations lorsque les tableaux ont la même forme donc possibilité de les supperposés
# (Critiquable...)

# Si on change B par exemple en (2, 2) on aura une erreur;

B = np.ones((2, 2))

A + B
````

````python
# Bah voilà on a une erreur, ce qui est parfaitement normal.
# Car nos tableaux ne peuvent se supperposer
````

````python
# Là vous vous dites qu'on ne peut faire ces opérations que sur des tableaux ayant mêmes dimensions, mais pas tout à
# faire...

# On peut aussi faire ces operations mais on va étendre les dimensions. Et c'est çà le broadcasting

# Le Broadcasting consiste à étendre les dimensions d'un tableau

# Il y'a une règle à suivre quand on fait du broadcasting

# La règle est simple:
# A et B dimensions égales ou égales à 1
````

````python
# Donc soit même nombre de lignes et même nombre de colonnes
# Ou soit 1

A = np.random.randint(0, 10, [2, 3])
B = np.ones((2, 1))
A
````

````python
B
````

````python
A + B
````

````python
#ceci marche | si B était (1, 2) çà n'allait pas marcher

B = np.ones((1, 2))
A + B
````

````python
# (1, 3) pour B va marcher par contre

B = np.ones((1, 3))

A + B
````

Quand on utilise numpy c'est pour travailler avec des tableaux de ndimensions.

Axe 0 : qui représente l'axe verticale (les lignes)
Axe 1 : qui représente l'axe horizontale (les colonnes)

Ce system d'axes est l'une des choses les plus importantes qu'il faut connaître quand on travaille avec numpy
parce que ça aura de l'impact sur la façon dont vous faîtes de l'indexing, slicing ou bien même sur toutes
les méthodes que vous pourriez utiliser.

Ensuite, concernant les attributs du tableau à ndimensions, on aura deux principalement à retenir:
    Le plus important de tout çà, est shape
    ensuite size

L'attribut shape nous permet d'analyser les dimensions de notre tableau, savoir combien d'éléments on a
sur axe 0 et sur axe 1

L'attribut size qui nous permet de connaître le nb d'éléments dans notre tableau.
