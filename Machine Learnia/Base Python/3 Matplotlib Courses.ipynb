## Customisation de Courbes

Les possibilités sont infinies
C'est bien le problèmes, donc faites simples

#### Quelques Paramètres

- label => Pour donner un nom à notre courbe
- c => Pour choisir la couleur de notre courbe "ex: red"
- lw (line width) => Pour choisir l'épaisseur de notre ligne en entier "ex: 3"
- ls (line style) => Pour choisir le style de notre ligne "ex: --"


#### La chose la plus importante c'est:

Le cycle de vie d'une figure.

Avant quand on voulait créer notre courbe on a commencé par plt.plot(x, y), c'est simple , c'est efficace
mais en réalité quand on veut créer une figure on commence ainsi:

```python
# Création d'un figure:

plt.figure() ## Ceci crée une feuille ou fenêtre de travail vide
```

```python
# Mtn on peut mettre notre courbe dessus
plt.figure()
plt.plot(x, y)
```

```python
# Mtn la raison pour laquelle on a explicitement mis plt.figure()
# C'est parce qu'à l'intérieur des paramètres on va pouvoir choisir la taille de notre figure
```

### Les paramètres pour la fonction figure(...)

- figsize(12, 8) ## Utile lorsqu'on a besoin d'avoir une figure grande sur laquelle vous voulez avoir beaucoup de détails dans notre figure

```python
plt.figure(figsize=(12, 8))
plt.plot(x, y)
plt.show()
```

```python
# Ensuite dans le cycle de vie d'une figure on peut continuer à ajouter d'autres courbes

plt.figure(figsize=(12, 8))
plt.plot(x, y)
plt.plot(x, x**3)
plt.show()
```

```python
# on peut ajouter quelques détails si on veut

plt.figure(figsize=(8, 5))
plt.plot(x, y)
plt.plot(x, x**3)
plt.title('figure 1') # Nom de la figure
plt.xlabel('axe x') # Nom de l'axe des abscisses
plt.ylabel('axe y') # Nom de l'axe des ordonnées
plt.show()
```

```python
# On peut aussi ajouter des legendes pour indiquer à quoi correspondent nos deux petites courbes
# Pour çà on va le faire dans les deux fonctions plot des deux courbes

plt.figure()
plt.plot(x, y, label='quadratique')
plt.plot(x, x**3, label='cubique')
plt.title('figure 1') # Nom de la figure
plt.xlabel('axe x') # Nom de l'axe des abscisses
plt.ylabel('axe y') # Nom de l'axe des ordonnées
plt.legend()
plt.show()
```

```python
# Pour finir une toute dernière chose qu'il faut avoir lorqu'on fait des graphiques, c'est de pouvoir les sauvegarder
# avec la fonciton plt.savefig()

plt.figure(figsize=(12, 10))
plt.plot(x, y, label='quadratique')
plt.plot(x, x**3, label='cubique')
plt.title('figure 1') # Nom de la figure
plt.xlabel('axe x') # Nom de l'axe des abscisses
plt.ylabel('axe y') # Nom de l'axe des ordonnées
plt.legend()
plt.show()
plt.savefig('figure.png')

plt.figure()
plt.plot(x, y)
```

```python
# Mtn des fois, on a envie d'afficher une grille de graphiques sur une seule et une seule figure.
# Pour çà, on utilise la fonction subplot
```

#### Subplot (lignes, colonnes, position)

```python
# Créons une grille de 1 colonnes et 2 ligne

plt.subplot(2, 1, 1) # 2 => lignes 1 => colonnes et 1 => la position de notre première graphique
plt.plot(x, y, c='red')
plt.show()
```

```python
# Mtn on va ajouter notre graphique qui se situe en dessous:

plt.subplot(2, 1, 1) # 2 => lignes 1 => colonnes et 1 => la position de notre première graphique
plt.plot(x, y, c='red')
plt.subplot(2, 1, 2) # 2 => lignes 1 => colonnes et 2 => la position de notre deuxième graphique
plt.plot(x, y, c='green')
plt.show()
```

## La méthode de OOP

```python
# On va commencer par créer notre figure et les axes avec la fonction subplots(s à la fin)

fig, ax = plt.subplots() ## fig, et ax sont des objets
#Cela voudrait dire qu'ils ont des méthodes et attributs qui vont permettre de les manipuler

# On va mtn tracer notre graphique avec la méthode plot de ax
ax.plot(x, y)
```

```python
# La différence avec la OOP c'est qu'on a plus de possibilités:
# On peut par exemple créer des graphiques qui partagent la même abscisse ou la même ordonnées
# Pour çà il faut créer plusieurs graphiques dans subplots et activé la fonctionnalité que l'on souhaite

fig, ax = plt.subplots(2, 1, sharex=True) # 2 lignes et 1 colonne et sharex => Pour dire qu'ils vont partager la même absicisse

ax.plot(x, y)
plt.show()
```

```python
# On a une erreur ! Pas de panique

# En fait quand on crée notre fig et ax dans ce cas 

# fig => est un objet
# ax => est un tableau numpy qui contient nos objets 0 => axe 1 et 1 => axe 2

# Donc ce qu'on va faire :

fig, ax = plt.subplots(2, 1, sharex=True) # 2 lignes et 1 colonne et sharex => Pour dire qu'ils vont partager la même absicisse

ax[0].plot(x, y, c="red")
ax[1].plot(x, np.sin(x**2), c="blue")
plt.show()
```

```python
# Et on voit bien que nos graphs partagent la même abscisse
```

## Exercise

Vous avez un dictionnaire qui comprend N Datasets

Créez une fonction qui trace N graphiques sur une seule figure

```python
dataset = {f"experience{i}": np.random.randn(100) for i in range(4)}
```

```python
def graphique(dataset):
    N = len(dataset)
    plt.figure(figsize=(12, 8))
    for i in range(N):
        x = dataset["experience{}".format(i)]
        plt.subplot(N, 1, i + 1)
        plt.plot(x)
    plt.show()
        
graphique(dataset)
```
