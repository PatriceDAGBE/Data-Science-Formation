## 5) SCATTER (Classification)

En Machine Learning, pratiquement la moitié des problèmes sur lesquels on travaille sont des problèmes
de classiffication.

Par exemple on désire classer un email en tant que SPAM et NON SPAM en fonction de plusieurs attributs que cet email 
va avoir comme le "nombre de fautes d'ortographes" ou le "nombre de liens internet présents".
Typiquement, dans ce problème on a deux classes: SPAM et NON SPAM

Et pour représenter un tel dataset, le graphique plt.scatter étant des meilleurs choix que vous pourriez faire.

======>

```python
### Faison un exemple avec le dataset connu [Iris flower data set]
```

```python
# Ce dataset contient 150 exemples de fleurs d'Iris reparti en trois (3) classes et dispose de quatre (4) variables
# pour prédire quelle classe il s'agit.

#===============================
#Classes:
#===============================
# Setosa
# Versicolor
# Virginicar

#===============================
#Variables:
#===============================
# Sepal length
# Sepal width
# Petal length
# Petal width
```

```python
# D'abord chargeons le dataset

#!pip install scikit-learn
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris() # Notre dataset
```

```python
x = iris.data # Ce data represente nos variables
y = iris.target # Ceci nos classes [0..2]
names = iris.target_names.tolist() # Les noms de nos classes
names
```

```python
x
```

```python
y
```

```python
## A présent faisons une visualization avec scatter

# On va passer deux de ces variables à scatter

plt.scatter(x[:, 0], x[:, 1])
```

```python
# On a un nuage de points
# Mais ce qui serait intéressant c'est de colorier ces points là suivant la nature de leur classe

plt.scatter(x[:, 0], x[:, 1], c=y)
plt.xlabel("Longueur du sépal")
plt.ylabel("Largeur du sépal")
plt.show()
```

```python
# Et là on a déjà un graphique super intéressant qui peut déjà nous apprendre beaucoup de choses sur le problème 
# que l'on a à résoudre

# Bien sûr l'une des limites de ce graphique c'est que l'on peut uniquement répresenter deux (2) de ces variables sur
# les quatre (4) dont on dispose ===> ON VERRA PLUS TARD COMMENT FAIRE (Technique)

# Pour finir avec le graphique scatter, voyons deux astuces personnelles:

# 1) La prémière c'est alpha
    #Il nous permet de contrôler la transparence de chacun des points de notre graph

plt.scatter(x[:, 0], x[:, 1], c=y, alpha=0.5)
plt.xlabel("Longueur du sépal")
plt.ylabel("Largeur du sépal")
plt.show()
```

```python
# 1) La prémière c'est s
    #Il nous permet de contrôler la taille de nos points (ou en fonction de nos variables => deuxième figure ci-dessous:)
    
plt.scatter(x[:, 0], x[:, 1], c=y, alpha=0.5, s=100)
plt.xlabel("Longueur du sépal")
plt.ylabel("Largeur du sépal")
plt.show()

plt.scatter(x[:, 0], x[:, 1], c=y, alpha=0.5, s=x[:, 2]*100)
plt.xlabel("Longueur du sépal")
plt.ylabel("Largeur du sépal")

plt.show()
```

```python
# Plus ils sont gros, plus le pétal est long et plus ils sont petits, plus le pétal est petit
```

## 4) GRAPHIQUE 3D

Ce type de graphique nous vient de mpl_toolkits.mplot3d qui est un des subpackages de matplotlib.
Donc vous l'aurez compris, avec çà on peut faire des graphiques en 3 dimensions, ce qui peut être super 
utile quand on désire visualiser plus de variables d'un seul coup sur un seul et même graphique.

```python
# Pour faire ce genre de chose on va commencer par le charger
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Une fois qu'on a chargé ce module, il va falloir créer un objet ax avec la fonction axes dans plt
# Et cet axe on va le faire en 3d en définissant projection sur 3d

ax = plt.axes(projection="3d")
```

```python
# Lorsqu'on crée des objets avec matplotlib on est obligé d'utiliser la OOP! Pas de panique...

# Pour çà on va créer notre scatter mais cette fois avec 3 variables

ax = plt.axes(projection="3d")
ax.scatter(x[:, 0], x[:, 1], x[:, 2], c=y)
plt.show()
```

```python
# Pour visualiser cet graph en 3d lorsque vous êtes dans jupyter notebook il faut decommenter %matplotlib ceci:
#%matplotlib
#ax = plt.axes(projection="3d")
#ax.scatter(x[:, 0], x[:, 1], x[:, 2], c=y)
#plt.show()
```

```python
# Ce qu'on pourrait également faire avec des graphiques 3d c'est de visualiser des surfaces et des fonctions
# mathématiques que l'on aurait modelisées. çà c'est super interessant.

# Créons alors une fonction avec lambda qui est notre générateur de fonctions anonymes

f = lambda x, y : np.sin(x) + np.cos(x+y)
```

```python
# Mtn si on veut observer f(x, y) en fonction de x et y , alors ce qu'il d'abord falloir faire c'est:

# - Créer deux vecteurs numpy x et y
# - Créer une sorte de grille en utilisant meshgrid [Nous permet de créer une grille à partir de deux axes]

X = np.linspace(0, 5, 100)
Y = np.linspace(0, 5, 100)
X, Y = np.meshgrid(X, Y)
```

```python
# On peut à présent injecter nos variables X et Y à l'intérieur de notre fonction f(x, y)

Z = f(X, Y)
```

```python
# Pour visualiser tout çà, on va faire recours à notre mpl_toolkits

ax = plt.axes(projection="3d")
ax.plot_surface(X, Y, Z, cmap="plasma") #cmap => pour choisir une couleur pour notre surface
plt.show()
```

## 3) HISTOGRAMME

Ce graphique est juste indispensable à connaître en Machine Learning et en Data Science.
Il nous permet de voir la distribution des données avec lesquelles on travail.

Il est important de savoir si nos données suivent une distribution normale ou non.
Ou bien si la distribution est symétrique ou asymétrique.
Ou bien de visualiser où se situe la moyenne...etc.

```python
# Par exemple pour reprendre notre dataset avec les fleur d'Iris, on peut visualiser la distribution de nos quatre (4)
# variables

# Pour çà, on va faire une des 4 variables à notre fonction hist

plt.hist(x[:, 0])
plt.show()
```

```python
# Ceci nous donne la distribution de la variable 0 de nos fleurs d'Iris
# Qu'est ce que ça represente vraiment ?

# Alors, sur l'axe des abscisses, on a toutes les valeurs que peut prendre la longueur d'un sépal dans notre fleur d'Iris
# Donc le sépal le plus petit qu'on a, il fait environ 4.5cm et le plus long fait environ 8cm
# Ensuite , ce qu'on a sur l'axe des ordonnées, c'est tout simplement le nombre d'apparition de chacune de ces
# catégories
# Donc par exemple dans notre dataset on environ 25 fois un sépal qui est aux environs de 5.5cm
```

```python
# ENSUITE
# Dans notre fonction hist on peut choisir de modifier un paramètre qui s'appelle bins
# qui détermine le nombre de sections que l'on désire avoir dans notre histogramme donc par défaut
# bins=10

plt.hist(x[:, 0], bins=30)
plt.show()
```

```python
# Là on a un histogramme beaucoup plus détaillé
# Si à l'inverse on créait un histogramme avec bins=1

plt.hist(x[:, 0], bins=1)
plt.show()
```

```python
# Là vous verrez que le nombre total qui tombe dans une seule catégorie est 150 => qui équivaut au nombre total de fleurs qu'on a
# dans notre dataset

# Mtn sur une seule et même figure, on peut créer deux histogrammes en même temps

plt.hist(x[:, 0], bins=20)
plt.hist(x[:, 1], bins=20)
plt.show()
```

```python
# Mais les graphiques avec histogramme dans matplotlib, ça ne s'arrête pas là
# Parce qu'on peut également tracer des histogrammes en 2d pour visualiser la distribution des données
# lorsqu'elles suivent deux (2) variables

# Et pour çà, on passe 2 variables à notre fonction hist2d...

plt.hist2d(x[:, 0], x[:, 1])
plt.xlabel("Longueur du sépal")
plt.ylabel("Largeur du sépal")
plt.show()
```

```python
# Pour bien comprendre, ajoutons une colorbar qui nous indique à quoi corresponde chaque couleur sur notre
# histogramme (une legende)

plt.hist2d(x[:, 0], x[:, 1])
plt.xlabel("Longueur du sépal")
plt.ylabel("Largeur du sépal")
plt.colorbar()
plt.show()
```

```python
# Jaune => indique les fréquences d'apparition les plus élévées
# ...
# Bleue pure => les moins élévées

# Donc là on voit qu'on obtient environ 10 sépals qui ont une longueur de sépal qui sont aux alentours de 6.5cm
# avec une largeur d'environ 3.5cm
# Donc on va vraiment trouver la majorité des fleurs d'Iris dans cette catégorie

# Il y'a une catégorie qui n'apparaît presque jamais c'est le carré tout en haut (bleu pur)
# qui est aux environs de 5.5cm de longueur avec une largeur d'environ 4.5cm

# Pour rendre tout ceci un peu plus attirant on peut leur ajouter une couleur

plt.hist2d(x[:, 0], x[:, 1], cmap="Blues")
plt.xlabel("Longueur du sépal")
plt.ylabel("Largeur du sépal")
plt.colorbar()
plt.show()
```

```python
# Et voilà ça nous donne d'autres couleurs en clair.
```

```python
# Pour finir sur les histogrammes, il y'a une dernière utilité super interessante, c'est lorsqu'on utilise
# pour analyser d'une image...

# Typiquement, quand on veut faire l'analyse d'une image en noir et blanc, on peut créer un histogramme
# dans lequel bins est égal à 255 parceque une image en général comprend 255 valeurs de pixels avec 0 répresentant
# le noir et 255 représentant le blanc.

# Importons d'abord notre image... Et on va l'enregistrer dans un tableau numpy qui s'appelle face
# que l'on va applatir avec ravel()
```

```python
from scipy import misc
import matplotlib.pyplot as plt
face = misc.face(gray=True)
plt.imshow(face, cmap="gray")
```

```python
plt.hist(face.ravel(), bins=255)
plt.show()
```

```python
# On peut ainsi en faire l'analyse en observant la fréquence d'apparition des différentes couleurs de pixels
# Typiquement, on voit qu'il y'a une majorité de pixels qui sont entre 100 et 150
# On a très peu de pixels qui sont très clairs
# Et pas mal de pixels qui sont dans le noir
```

## 2) CONTOUR PLOT

Ces graphiques sont super importantes quand l'on veut visualiser des models qui occupent trois dimensions
en vue du dessus.

Souvent en Intelligence Artificielle, on cherche souvent à résoudre des problèmes d'optimisation comme par
exemple maximiser des revenus etc...

Eh bien ce type de graphique peut jouer un rôle majeur dans votre compréhension du problème d'optimisation dans
lequel vous travaillez.

```python
# Typiquement, tout à l'heure on a créé une fonction f = lambda x, y : np.sin(x) + np.cos(x+y)
# Et on l'avait visualisé en 3d, alors si on veut avec contour plot on peut la visualiser en vue de
# dessus
```

```python
plt.contour(X, Y, Z)
```

```python
# Pour rendre cette graphique un petit peu plus lisible alors on peut en modifier le nombre de niveaux
# 20 par exemple qui va augmenter la précision d'affichage qu'on a sur notre graphique.

plt.contour(X, Y, Z, 20)
```

```python
# Si on veut on peut l'afficher en noir et blanc...

plt.contour(X, Y, Z, 20, colors="black")
```

```python
# Là tout ce qui est positif s'affiche en continu et tout ce qui est négatif en pointillé

# Donc typiquement si vous travaillez en économie, ingénierie, en logistique, en fait quelque soit le secteur et 
# que vous avez une fonction un peu plus complexe à optimiser et bien les coutour plot peuveut vraiment vous aider à y voir clair
# dans un champ de possibilité.

# Ex: 
f = lambda x, y : np.sin(x) + np.cos(x+y)*np.cos(x)
X = np.linspace(0, 5, 100)
Y = np.linspace(0, 5, 100)
X, Y = np.meshgrid(X, Y)
Z = f(X, Y)

plt.contour(X, Y, Z, 20, colors="black")
```

```python
# Par exemple ici on voudra attendre le maximum de notre fonction en visant le rond point en continu ou le minimum
# en visant le rond point en pointillé

# Mtn petite astuce pour les contour plot:
# Il existe une autre fonction contourf dans laquelle on aussi faire passer X Y et Z et également cmap (color map)

plt.contourf(X, Y, Z, 20, cmap="RdGy")
plt.colorbar()
```

```python
# Et là ça nous donne une graphique un petit peu plus sympa.
# Avec çà on peut comprendre quelles sont les valeurs de nos différents gradients de notre contour plots
# Très très utile...
```

## 1) IMSHOW

Imshow c'est sans contexte, le graphique à la tête de ce classement...

Bah, bien sûr la toute première utilité c'est de visualiser une image... comme l'image face que l'on avait chargée
auparavant.

```python
plt.imshow(face)
```

Mais en fait l'utilisation de imshow() ça va beaucoup plus loin que simplement afficher des images...

Rappelez vous d'une chose, face c'est un tableau numpy qui contient les différents pixels de notre image

Donc en fait la fonction imshow() elle nous permet d'afficher n'importe quel tableau numpy et c'est l'intérêt qui est génial qu'on a à utiliser imshow.

On va pouvoir afficher n'importe quelle matrice, par exemple on va pouvoir tracer des matrices de corrélation , on 
va pouvoir analyser un problème d'optimisation en mettant tous les chemins possibles dans une matrice, on va pouvoir visualiser un mask avant de s'en servir dans du booléan indexing, bref on peut énormément de choses avec imshow()



```python
# Par exemple avec imshow on peut tracer la matrice de corrélation pour notre dataset de fleus d'Iris

# Donc pour tracer un tel graphique, on va faire passer à l'intérieur de imshow la fonction np.corrcoef() qui nous
# retourne une matrice de corrélation pour notre tableau x...

# Donc on va faire passer la transposée de x parce que ce qui nous intéresse c'est la corrélation en les colonnes
# non les lignes

plt.imshow(np.corrcoef(x.T), cmap="Blues")
plt.colorbar()
plt.show()
```

```python
# Et çà nous produit la matrice de corrélation qui nous permet de voir où se situent les différentes corrélations
# dans notre dataset...

# Mtn quand on travaille sur des dataset beaucoup plus larges comme par exemple le dataset sur les données du
# cancer de seins (Breast Cancer) et qu'on a beaucoup plus de variables (ex:30) et bien ce genre de 
# graphique est super informatif pour comprendre quelles sont les variables qui ont de fortes corrélations
# les unes avec les autres

# Voilà on pourrait encore parler très longtemps de imshow() mais on va simplement parler des problèmes d'optimisation
# pour lesquels avant on utilisait contour plot et bien tout ce qu'on a à faire c'est de passer notre tableau Z
# qui est donc un tableau de ndim 100 lignes par 100 colonnes à l'interrieur de imshow()

plt.imshow(Z)
```
