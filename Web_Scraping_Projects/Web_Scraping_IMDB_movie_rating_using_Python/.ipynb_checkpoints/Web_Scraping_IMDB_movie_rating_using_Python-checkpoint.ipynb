```python
import requests
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import pandas as pd
import json
```

```python
url = "https://www.imdb.com/chart/top/"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
}

driver = webdriver.Chrome()
driver.get(url)

# Attendre que la page se charge
time.sleep(3)

# Récupérer le HTML
html = driver.page_source
soup = BeautifulSoup(html, "html.parser")
```

```python
movies = soup.select(".ipc-metadata-list li")
crew = [li.find(class_="ipc-title").get_text() for li in soup.select(".ipc-metadata-list li")]
ratings = [b.find(class_="ipc-rating-star--rating").get_text()
            for b in soup.select(".cli-ratings-container")]
ages_metadata = soup.select(".cli-title-metadata")

ages = [div.find(class_="cli-title-metadata-item").get_text() for div in ages_metadata]
```

```python
titles = [title.split('. ', 1)[1] for title in crew]
titles
```

```python
ratings
```

```python
list = []
```

```python
movie_soup_list = []
for movie_idx in range(len(movies)):
    movie_href = movies[movie_idx].find(class_="ipc-lockup-overlay").attrs.get("href")
    movie_url = "http://www.imdb.com" + movie_href
    
    driver.get(movie_url)
    
    #resp = requests.get(movie_url, headers=headers)
    resp = driver.page_source
    
    movie_soup = BeautifulSoup(resp, "html.parser")
    
    movie_soup_list.append(movie_soup)

print("Movies scraping done!")
```

```python
driver.quit()
```

```python
list = []

for idx, movie_soup in enumerate(movie_soup_list):
    
    description_tag = movie_soup.find("span", attrs={"data-testid": "plot-xl"})
    
    description = description_tag.get_text(strip=True) if description_tag else "Description non disponible"

    # === Crew (Director, Writers, Stars) ===
    crew_section = movie_soup.find_all("li", attrs={"data-testid": "title-pc-principal-credit"})
    crew_data = {}

    for li in crew_section:
        # Titre de rôle (peut être <span> ou <a>)
        role_tag = li.find("span", class_="ipc-metadata-list-item__label")
        if not role_tag:
            role_tag = li.find("a", class_="ipc-metadata-list-item__label")
        
        role_name = role_tag.get_text(strip=True) if role_tag else "Unknown Role"

        # Noms associés
        people = [a.get_text(strip=True) for a in li.find_all("a", class_="ipc-metadata-list-item__list-content-item")]
        crew_data[role_name] = people
    
    # === Données du film ===
    data = {
        "title": titles[idx],
        "description": description,
        "year": ages[idx],
        "rating": float(ratings[idx]),
        "director": crew_data.get("Director", []),
        "writers": crew_data.get("Writers", []),
        "stars": crew_data.get("Stars", [])
    }
    
    list.append(data)

list
```

```python
df = pd.DataFrame(list)
df.to_csv('imdb_top_250_movies.csv',index=False)
```
